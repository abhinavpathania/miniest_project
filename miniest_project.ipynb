```python
file_csv.columns=file_csv.columns.str.lower()
file_csv.columns
```

```python
file_csv['sale_price']=file_csv['list_price']-file_csv['discount']
```

```python
file_csv['profit']=file_csv['sale_price']-file_csv['cost_price']
file_csv.head()
```
